```python
def categ(row):
    if row['Fator_risco']<=10:
        return 1
    elif row['Fator_risco']<=20:
        return 2
    elif row['Fator_risco']<=30:
        return 3
    elif row['Fator_risco']<=40:
        return 4
    elif row['Fator_risco']<=50:
        return 5
    elif row['Fator_risco']<=60:
        return 6
    elif row['Fator_risco']<=70:
        return 7
    elif row['Fator_risco']<=80:
        return 8
    elif row['Fator_risco']<=90:
        return 9
    else:
        return 10
```

```python
df['Fator_risco'].values
#df['Fator_risco'] = categ(df.values)
```

```python
@vectorize(nopython=True, cache=True)
def categ1(i):
    
    if i<= 10.0:return 1
    elif i<=20.0:return 2
    elif i<=30.0:return 3
    elif i<=40.0:return 4
    elif i<=50.0:return 5
    elif i<=60.0:return 6
    elif i<=70.0:return 7
    elif i<=80.0:return 8
    elif i<=90.0:return 9
    else:return 10
```

```python
%%time
df['Fator_risco'] = categ1(df['Fator_risco'].values)
```

```python
X = df.drop('Anomalia', axis = 1)
```

```python
X
```

```python
%%time
iforest = IsolationForest(n_estimators=200, max_samples=0.1, contamination=0.05, max_features=1.0, bootstrap=False, n_jobs=-1, 
                            random_state=None, verbose=0, warm_start=False).fit(X)
```

```python
results = iforest.predict(X)
```

```python
results = np.where(results == 1, 0, 1)
```

```python
from sklearn.metrics import classification_report
print(classification_report(results, df['Anomalia']))
```
