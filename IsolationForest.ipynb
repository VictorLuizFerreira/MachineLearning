```python
from sklearn.preprocessing import OrdinalEncoder

enc = OrdinalEncoder()
forest[["Nome","Maquina", "IP"]] = enc.fit_transform(forest[["Nome","Maquina", "IP"]])
```

```python
forest
```

```python
def sla(x):
    if x > 10:
        return 1
    elif x > 20:
        return 2
    elif x > 30:
        return 3
    elif x > 40:
        return 4
    elif x > 50:
        return 5
    elif x > 60:
        return 6
    elif x > 70:
        return 7
    elif x > 80:
        return 8
    elif x > 90:
        return 9
    else:
        return 10
```

```python
forest['Fator_risco'] = sla(forest['Fator_risco'].values)
```
