```python
import pandas as pd
import numpy as np
import bz2
import _pickle as cPickle
import pickle
```

```python
def decompress_pickle(file):
    data = bz2.BZ2File(file, 'rb')
    data = cPickle.load(data)
    return data
```

```python
def compressed_pickle(title, data):
    with bz2.BZ2File(title + '.pbz2', 'w') as f:
        cPickle.dump(data, f)
    return data
```
